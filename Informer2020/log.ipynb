```python
import re
from pathlib import Path

folder_path =  Path('log') # ログファイルのパスを指定
file_names =  [file.name for file in folder_path.iterdir() if file.is_file()]
print(file_names)
mse_values = []
file_check = []

for file_name in file_names:
    log_file_path = "log\\"+file_name
    with open(log_file_path, 'r', encoding='utf-16') as file:
        for line in file:
            if "mse:" in line:
                # "mse:"の後の数値を抽出
                mse_value = float(line.split("mse:")[1].split(",")[0].strip())
                mse_values.append(mse_value)
                file_check.append(file_name)
print(mse_values)
print(file_check)
print(len(mse_values))
print(len(file_check))

min_value = min(mse_values)
min_index = mse_values.index(min_value)
print("配列の最小値は:", min_value)
print("最小値のインデックスは:", min_index)
```
